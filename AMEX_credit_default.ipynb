```python
keys = []
for column in data.columns:
  if len(data[column].unique())== data.shape[0]:
    keys.append(column)
```

```python
len(keys)
```

```python
benchmark = data.drop(keys,axis=1)
benchmark.shape
```

```python
len(benchmark.customer_ID[benchmark.D_87==1].unique())
```

```python
temp =pd.DataFrame(benchmark.isna().sum())
sorteduniqcnt = temp.sort_values(0,ascending=False)
sorteduniqcnt
```

```python
#Create list of columns to be analysed
undinv = sorteduniqcnt[sorteduniqcnt[0] >= 0.50*benchmark.shape[0]].index.values
```

```python

for xname in undinv:
    #print(xname + "\n"+ str(benchmark[benchmark.xname.notna()].target.value_counts()))
    title = str(benchmark[benchmark[xname].notna()].target.value_counts())
    sns.displot( benchmark, x=xname, hue="target", kind="kde",warn_singular=False).set(title=title)
    sns.displot(benchmark, x=xname, hue="target", multiple="dodge").set(title=title)

```
