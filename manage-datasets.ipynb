```python
import sklearn as skl
import dask as dask
import dask.dataframe as dd
from dask_ml import model_selection
import random
```

```python
dataset = dd.read_csv("/kaggle/input/amex-default-prediction/train_data.csv")
labels = dd.read_csv("/kaggle/input/amex-default-prediction/train_labels.csv")
```

```python
dict = {"B":0, "R":0, "D":0,"S":0, "P":0, "Others":0}
for col in sorted(dataset.columns):
    start_letter = col[0]
    if start_letter in dict.keys():
        dict[start_letter]+=1
    else:
        dict["Others"]+=1   
print(dict)
```

```python
random.seed(0)
train_dataset = dataset.sample(frac=0.05, replace=False, random_state=3)
random.seed(0)
test_dataset = labels.sample(frac=0.05, replace=False, random_state=3)
```

```python
train_dataset.shape[0].compute()
```

```python
train_dataset.compute().to_csv("sample_train_one.csv", index=False)
```

```python
y = test_dataset.compute()
test_dataset.to_csv("sample_test_one.csv", index=False)
```

```python
combined_dataset = dask.dataframe.multi.merge(left = dataset, right=labels, how="inner",on="customer_ID")
```

```python
computed_combined_dataset = combined_dataset.sample(frac=0.05, replace=False, random_state=3).compute()
```

```python
computed_combined_dataset.to_csv("combined_sample.csv", index=False)
```
